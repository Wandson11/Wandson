```python
Dados['Amarela'] = pd.to_numeric(Dados['Amarela'])
```

```python
Dados['Indígena'] = pd.to_numeric(Dados['Indígena'])
```

```python
Dados.info()
```

```python
Dados['Preta'] = pd.to_numeric(Dados['Preta'].apply(lambda linha: linha.replace(',', '')))
```

```python
Dados['Parda'] = pd.to_numeric(Dados['Parda'].apply(lambda linha: linha.replace(',', '')))
Dados['Não Informado'] = pd.to_numeric(Dados['Não Informado'].apply(lambda linha: linha.replace(',', '')))
```

```python
#ideia agora é a de apurar o total por mes e por coluna, ou seja, por raça
#para esse objetivo, usarei sum()
Colunas = Dados.sum(axis=0).reset_index()
Colunas
#inserido axis=0, a soma será feita por coluna
```

```python
#agora contagem de total por mes (linha)
linhas = Dados.sum(axis=1).reset_index()
linhas
```

```python
#janeiro é o mes com mais frequencia, enquanto dezembro está no extremo oposto.
linhas.rename(columns={
    'index': 'Meses',
    0: 'Mortes'
}, inplace=True)
linhas
```

```python
plt.figure(figsize=(15,6))
plt.title('Crimes por meses')
sns.lineplot(linhas['Meses'], linhas['Mortes'])
```

```python
total_linhas = linhas['Mortes'].sum()
total_linhas
```

```python
#aplicando o sum sob as linhas e colunas, tem-se o resultado inicial, agora falta dividir
```

```python
linhas['%'] = (linhas['Mortes']/total_linhas) * 100
linhas
```

```python
Colunas['%'] = (Colunas['Frequencia']/total_linhas) * 100
Colunas
```

```python
#com esse cálculo, nota-se que menos de 1%, separa a cor parda (lider de mais suspeito) para a cor branca, 35,42% e 34,90% respectivamente
```
