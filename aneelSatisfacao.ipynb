```python
plt.figure(figsize=(17,8))
plt.style.use('seaborn-darkgrid')
sns.barplot(dados['SigAgente'],dados['MdaIndicadorQualidade'], ci=None) #ci=None, tira os risquinhos que aparecem em grafico de barra
plt.xticks(rotation=90)
plt.show()
```

```python
dados['MdaIndicadorQualidade'].sort_values(ascending=False)
```

```python
top10 = dados.groupby(dados['SigAgente']).mean()['MdaIndicadorQualidade']
top10
```

```python
lista10=top10.sort_values(ascending=False).head(10).reset_index()
lista10
```

```python
plt.figure(figsize=(15,6))
sns.barplot(lista10['SigAgente'], lista10['MdaIndicadorQualidade'], color='blue')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(15,6))
sns.scatterplot(dados['SigAgente'], dados['MdaIndicadorQualidade'])
plt.xticks(rotation=90)
plt.show()
```
