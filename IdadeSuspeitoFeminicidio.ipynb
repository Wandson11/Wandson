```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dados = pd.read_csv('/content/Balanco-2019-Ligue180-Idade Suspeito.csv')
Dados.shape
```

```python
Dados.info()
```

```python
Dados.drop(columns=['Unnamed: 0', 'Unnamed: 14'], inplace=True)
```

```python
Dados.drop(index=[0, 1, 2, 3, 4, 5, 6, 7,26, 27, 28, 29], inplace=True)
```

```python
Dados
```

```python
Dados.columns
```

```python
Dados.rename(columns={
    'Unnamed: 1': 'Faixa Etária - Suspeito',
    'Unnamed: 2': 'Jan',
    'Unnamed: 3': 'Fev',
    'Unnamed: 4': 'Mar',
    'Unnamed: 5': 'Abr',
    'Unnamed: 6': 'Mai',
    'Unnamed: 7': 'Jun',
    'Unnamed: 8': 'Jul',
    'Unnamed: 9': 'Aug',
    'Unnamed: 10': 'Set',
    'Unnamed: 11': 'Out',
    'Unnamed: 12': 'Nov',
    'Unnamed: 13': 'Dez'
}, inplace=True)
```

```python
Dados.head()
```

```python
Dados.drop(index=8, inplace=True)
```

```python
Dados.set_index('Faixa Etária - Suspeito', inplace=True)
```

```python
Dados = Dados.T
Dados
```

```python
Dados['0 a 3 anos'] = pd.to_numeric(Dados['0 a 3 anos'])
Dados['4 a 7 anos'] = pd.to_numeric(Dados['4 a 7 anos'])
Dados['8 a 11 anos'] = pd.to_numeric(Dados['8 a 11 anos'])
Dados['12 a 14 anos'] = pd.to_numeric(Dados['12 a 14 anos'])
Dados['15 a 17 anos'] = pd.to_numeric(Dados['15 a 17 anos'])
Dados['18 a 24 anos'] = pd.to_numeric(Dados['18 a 24 anos'])
Dados['25 a 30 anos'] = pd.to_numeric(Dados['25 a 30 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['31 a 35 anos'] = pd.to_numeric(Dados['31 a 35 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['36 a 40 anos'] = pd.to_numeric(Dados['36 a 40 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['41 a 45 anos'] = pd.to_numeric(Dados['41 a 45 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['46 a 50 anos'] = pd.to_numeric(Dados['46 a 50 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['51 a 55 anos'] = pd.to_numeric(Dados['51 a 55 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['56 a 60 anos'] = pd.to_numeric(Dados['56 a 60 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['61 a 65 anos'] = pd.to_numeric(Dados['61 a 65 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['66 a 70 anos'] = pd.to_numeric(Dados['66 a 70 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['Maior de 70 anos'] = pd.to_numeric(Dados['Maior de 70 anos'].apply(lambda linha1: linha1.replace(',', '')))
Dados['Não Informado'] = pd.to_numeric(Dados['Não Informado'].apply(lambda linha1: linha1.replace(',', '')))
```

```python
#no conjunto de dados acima, inciailmente foram transformados em float, porém, surgiu um problema, onde valores inferioes a mim, depois da transformação viraram milhares
#por exemplo, em um mes tinha um valor de 990, com a transformação do repalce de , para ., esse valor virou 990.000, ficando assim, totalmente errado
#depois de alguns testes, resolvi transformar a virgula em nada, ficando assim, todos os valores inteiros.
#o interprador soube reconhecer e foi possivel montar os gráficos abaixos.
Dados.info()
```

```python
#tentando agrupar
Dados.groupby(['0 a 3 anos']).sum().reset_index()
#por ausencia de uma coluna preenchida com objet, utilizei a "0 a 3 anos" para apresentar a soma total de cada grupo etário
```

```python
#soma total de vitimas de feminicio por mês (linha)
Soma_Mes = Dados.sum(axis=1).reset_index()
Soma_Mes
```

```python
plt.figure(figsize=(15,6))
plt.style.use('seaborn-darkgrid')
plt.title('Vitimas de feminicidio - 2019')
sns.lineplot(Soma_Mes['index'], Soma_Mes[0])
plt.xlabel('Período')
plt.ylabel('Quantidade de Mortes');
```

```python
Soma_Mes.rename(columns={
    'index': 'Mês',
    0: 'Total'
}, inplace=True)
```

```python
Soma_Mes
```

```python
#somando as colunas
Soma_Coluna = Dados.sum(axis=0).reset_index()
Soma_Coluna
```

```python
Soma_Coluna.rename(columns={
    'Faixa Etária - Suspeito': 'Idade Suspeito',
    0 : 'Vítimas Feminicídio'
}, inplace=True)
```

```python
plt.figure(figsize=(15,5))
plt.title('Vitimas de feminicidio por meses - 2019', fontsize=14)
sns.lineplot(Soma_Coluna['Idade Suspeito'], Soma_Coluna['Vítimas Feminicídio'])
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(15,5))
plt.title('Vítimas x Idade dos suspeitos', fontsize=14)
sns.barplot(Soma_Coluna['Idade Suspeito'], Soma_Coluna['Vítimas Feminicídio'])
plt.xticks(rotation=90)
plt.show()
```
