```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dados = pd.read_csv('/content/Balanco-2019-Ligue180-Suspeito.csv')
```

```python
Dados.shape
```

```python
Dados.tail(15)
```

```python
Dados.drop([0,1,2,3,4,5,6,7,8,57,58,59,60], inplace=True)
```

```python
Dados.drop(columns=['Unnamed: 0', 'Unnamed: 14'], inplace=True)
```

```python
Dados
```

```python
Dados.columns
```

```python
Dados.rename(columns={
    'Unnamed: 1': 'Relação Vítima x Suspeito', 
    'Unnamed: 2': 'Jan', 
    'Unnamed: 3': 'Fev', 
    'Unnamed: 4': 'Mar', 
    'Unnamed: 5': 'Abr',
    'Unnamed: 6': 'Mai', 
    'Unnamed: 7': 'Jun', 
    'Unnamed: 8': 'Jul', 
    'Unnamed: 9': 'Ago', 
    'Unnamed: 10': 'Set',
    'Unnamed: 11': 'Out', 
    'Unnamed: 12': 'Nov', 
    'Unnamed: 13': 'Dez'
}, inplace=True)
```

```python
Dados.shape
```

```python
Dados = Dados.set_index('Relação Vítima x Suspeito')
Dados
```

```python
Dados.info()
```

```python

Dados['Mar'] = pd.to_numeric(Dados['Mar'].apply(lambda linha: linha.replace(',', '')))
Dados['Abr'] = pd.to_numeric(Dados['Abr'].apply(lambda linha: linha.replace(',', '')))
Dados['Mai'] = pd.to_numeric(Dados['Mai'].apply(lambda linha: linha.replace(',', '')))
Dados['Jun'] = pd.to_numeric(Dados['Jun'].apply(lambda linha: linha.replace(',', '')))
Dados['Jul'] = pd.to_numeric(Dados['Jul'].apply(lambda linha: linha.replace(',', '')))
Dados['Ago'] = pd.to_numeric(Dados['Ago'].apply(lambda linha: linha.replace(',', '')))
Dados['Set'] = pd.to_numeric(Dados['Set'].apply(lambda linha: linha.replace(',', '')))
Dados['Out'] = pd.to_numeric(Dados['Out'].apply(lambda linha: linha.replace(',', '')))
Dados['Nov'] = pd.to_numeric(Dados['Nov'].apply(lambda linha: linha.replace(',', '')))
Dados['Dez'] = pd.to_numeric(Dados['Dez'].apply(lambda linha: linha.replace(',', '')))
```
