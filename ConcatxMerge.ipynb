```python
import pandas as pd
```

# Merge

```python
tabela1 = pd.DataFrame({
    'Nome': ['João', 'João', 'Pedro' , 'Caio'],
    'Telefone': ['12121', '343434', '565656', '787878'],
    'Carro': ['azul', 'preto', 'verde' , 'amarelo']
})
```

```python
tabela1
```

```python
tabela2 = pd.DataFrame({
    'Nome': ['João', 'Marcelo', 'Thiago' , 'Caio'],
    'Irmãos':['1', '3', '2' , '2']
})
tabela2
```

```python
#tipos de interseção entre as tabelas usando o merge, ou o procv do pandas.
#primeiro how = 'inner' (interseção exata entre tabelas, ou seja, apenas os nomes)
m = pd.merge(tabela1, tabela2, how='inner', on='Nome') #basicamente, foi dado como parametro os 2 df, com o tipo de interseção "inner" e qual, seria a coluna objeto, "nome"
m
#foi especificado "Nome", pois é a única coluna que se repete nos 2 df
```

```python
#segundo how = 'outer', informa que deseja todas as informações de ambos os df.
m = pd.merge(tabela1, tabela2, how='outer')
m
#é a mesma coisa do "inner" mas agora incluiu o "pedro", que nao tem irmão
#Marcelo e thiago não possuem dados de telefone e nem carro
```

```python
#terceiro how="left" ou "right", depende de qual tabela será deixada no lado direito ou esquerdo
m = pd.merge(tabela1, tabela2, how='left', on='Nome')
m
#somente ‘Nomes’ que existam na tabela_1 serão trazidos da tabela_2.
```

```python
m = pd.merge(tabela1, tabela2, how='right', on='Nome')
m
#aqui, foi puxada todas as informações, menos para pedro que nao tem irmaos.
```
