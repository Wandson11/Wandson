```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Base = pd.read_csv('/content/STP-comparativo.csv', encoding='latin-1', sep=';')
Base.head()
```

```python
Base.drop(index=(134), inplace=True)
```

```python
Base['Data'] = pd.to_datetime(Base['Data'])
```

```python
Base.head()
```

```python
Base.set_index(Base['Data'], inplace=True)
```

```python
Base.head()
```

```python
Base.rename(columns={
    '25434 - Taxa média mensal de juros das operações de crédito - Pessoas jurídicas - Total - % a.m.' : 'Jrs. Empresa',
    '25435 - Taxa média mensal de juros das operações de crédito - Pessoas físicas - Total - % a.m.' : 'Jrs. Físico'
}, inplace=True)
```

```python
Base.head()
```

```python
Base.drop(columns=('Data'), inplace=True)
```

```python
Base.head()
```

```python
Base.info()
```

```python
Base['Jrs. Empresa'] = pd.to_numeric(Base['Jrs. Empresa'].apply(lambda linha: linha.replace(',', '.')))
```

```python
Base['Jrs. Físico'] = pd.to_numeric(Base['Jrs. Físico'].apply(lambda linha1: linha1.replace(',', '.')))
```

```python
Base.head()
```

```python
Base.info()
```

```python
fig, ax=plt.subplots(figsize=(18,7))

plt.style.use('seaborn')
cor_fundo = '#f5f5f5'
ax.set_facecolor(cor_fundo)
fig.set_facecolor(cor_fundo)

plt.suptitle('Comparativo de gráficos plt com sns')

#PARAMETROS PARA O GRID
Linhas = 2
Colunas = 2

plt.subplot(Linhas, Colunas, 1)
plt.title('Gráfico de linha matplotlib')
plt.plot(Base.index, Base['Jrs. Empresa'])
plt.plot(Base.index, Base['Jrs. Físico'])

plt.xlabel('Período')
plt.ylabel('Juros')

plt.legend(['Juros Empresa', 'Juros Pessoa Física'])

plt.subplot(Linhas, Colunas, 2)
plt.title('Gráfico de linha seaborn')
sns.lineplot(Base.index, Base['Jrs. Empresa'])
sns.lineplot(Base.index, Base['Jrs. Físico'])

plt.xlabel('Período')
plt.ylabel('Juros')

plt.legend(['Juros Empresa', 'Juros Pessoa Física'], ncol=2, bbox_to_anchor=(0.75, -0.2))

plt.subplot(Linhas, Colunas, 3)
plt.title('Gráfico de barra do matplotlib')
plt.bar(Base.index, Base['Jrs. Empresa'])


plt.subplot(Linhas, Colunas, 4)
plt.title('Gráfico de barra seaborn')
sns.barplot(Base.index, Base['Jrs. Empresa'])
plt.xticks(rotation=90)

#ajustando o gráficos
plt.subplots_adjust(hspace=0.55, wspace=0.15)
plt.show()
```
