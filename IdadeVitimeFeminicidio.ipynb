```python
Dados = Dados.set_index('Faixa Etária - Vítima')
Dados
```

```python
Dados['0 a 3 anos'] = pd.to_numeric(Dados['0 a 3 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['4 a 7 anos'] = pd.to_numeric(Dados['4 a 7 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['8 a 11 anos'] = pd.to_numeric(Dados['8 a 11 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['12 a 14 anos'] = pd.to_numeric(Dados['12 a 14 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['15 a 17 anos'] = pd.to_numeric(Dados['15 a 17 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['18 a 24 anos'] = pd.to_numeric(Dados['18 a 24 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['25 a 30 anos'] = pd.to_numeric(Dados['25 a 30 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['31 a 35 anos'] = pd.to_numeric(Dados['31 a 35 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['36 a 40 anos'] = pd.to_numeric(Dados['36 a 40 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['41 a 45 anos'] = pd.to_numeric(Dados['41 a 45 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['46 a 50 anos'] = pd.to_numeric(Dados['46 a 50 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['51 a 55 anos'] = pd.to_numeric(Dados['51 a 55 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['56 a 60 anos'] = pd.to_numeric(Dados['56 a 60 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['61 a 65 anos'] = pd.to_numeric(Dados['61 a 65 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['66 a 70 anos'] = pd.to_numeric(Dados['66 a 70 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['Maior de 70 anos'] = pd.to_numeric(Dados['Maior de 70 anos'].apply(lambda linha: linha.replace(',', '')))
Dados['Não Informado'] = pd.to_numeric(Dados['Não Informado'].apply(lambda linha: linha.replace(',', '')))
```

```python
plt.figure(figsize=(15,5))
plt.title('Crianças vítimas de feminicidio', fontsize=14, loc='left')
plt.plot(Dados.index, Dados['0 a 3 anos'])
plt.plot(Dados.index, Dados['4 a 7 anos'])
plt.plot(Dados.index, Dados['8 a 11 anos'])

plt.legend(['0 a 3 anos', '4 a 7 anos', '8 a 11 anos'])

plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.style.use('seaborn-darkgrid')
plt.title('Adolescentes vítimas de feminicidios')
plt.plot(Dados.index, Dados['12 a 14 anos'])
plt.plot(Dados.index, Dados['15 a 17 anos'])

plt.legend(['12 a 14 anos', '15 a 17 anos'], ncol=2, bbox_to_anchor=(0.6, -0.1))
```
